```python
# Function to draw grid
def grid(image, rows=6, cols=8):
    grid_image = image.copy()
    height, width = image.shape[:2]  # get the first two as height & width 
    
    # calculate height & width（turn into int）
    tile_width = int(width / cols)
    tile_height = int(height / rows)
    
    # draw vertical grid（left to right）
    for x in range(0, width + 1, tile_width):  #  +1 make sure to draw the last line
        cv2.line(grid_image, (x, 0), (x, height), (0, 0, 0), 1)  # black grid
    
    # draw horizontal line（top to bottom）
    for y in range(0, height + 1, tile_height):  # +1 make sure to draw the last line
        cv2.line(grid_image, (0, y), (width, y), (0, 0, 0), 1)  # black grid
    
    return grid_image
```

```python
image = grid(starting_cut)
plt.imshow(image)
```

```python
# draw grid option 2
def draw_grid(image, grid_size):
    # Create a copy of the image to draw the grid
    grid_image = image.copy()
    height, width, _ = grid_image.shape

    # Vertical lines
    for x in range(0, width, grid_size):
        cv2.line(grid_image, (x, 0), (x, height), (0, 0, 0), 1)  # black lines (0,0,225) is red line

    # Horizontal lines
    for y in range(0, height, grid_size):
        cv2.line(grid_image, (0, y), (width, y), (0, 0, 0), 1)  # Red lines

    return grid_image
```

```python
def get_all_pixels_in_grid(image, rows=6, cols=8):
    height, width = image.shape[:2]
    tile_width = int(width / cols)
    tile_height = int(height / rows)
    
    # 初始化結果陣列
    grid_pixels = np.zeros((rows, cols, tile_height, tile_width, 3), dtype=np.uint8)
    
    for i in range(rows):
        for j in range(cols):
            x_start = j * tile_width
            x_end = (j + 1) * tile_width
            y_start = i * tile_height
            y_end = (i + 1) * tile_height
            
            grid_pixels[i, j] = image[y_start:y_end, x_start:x_end]
    
    return grid_pixels
```

## get each pixal unique color run def first then use pixal_color[x][y]

```python
def get_unique_colors_per_grid(all_pixels):
    rows, cols = all_pixels.shape[0], all_pixels.shape[1]
    unique_colors_per_grid = []  # 儲存每個 grid 的不重複顏色
    
    for i in range(rows):
        row_unique_colors = []
        for j in range(cols):
            # 取得當前 grid 的所有像素 (tile_height, tile_width, 3)
            grid_pixels = all_pixels[i, j]
            
            # 攤平像素並找出不重複顏色
            unique_colors = np.unique(
                grid_pixels.reshape(-1, 3),  # 轉換成 (N, 3) 形式
                axis=0
            )
            row_unique_colors.append(unique_colors)
        
        unique_colors_per_grid.append(row_unique_colors)
    
    # 轉換成 NumPy 陣列 (由於每個 grid 的 unique_colors 數量不同，無法直接轉成固定形狀的陣列)
    # 這裡我們返回一個 List of Lists，每個元素是 (N, 3) 的陣列
    return unique_colors_per_grid
```

```python
# 定義目標顏色（BGR 格式，因為 OpenCV 使用 BGR）
target_colors = [
    np.array([161, 130, 105]),  # [161, 130, 105] 的 BGR 格式
    np.array([203, 187, 168])   # [203, 187, 168] 的 BGR 格式
]

def draw_grid_borders(image, all_pixels, rows=6, cols=8):
    height, width = image.shape[:2]
    tile_width = width // cols
    tile_height = height // rows
    
    for i in range(rows):
        for j in range(cols):
            # 取得當前 grid 的所有像素
            grid_pixels = all_pixels[i, j]
            
            # 檢查是否所有像素都是目標顏色之一
            is_target_color = True
            for pixel in grid_pixels.reshape(-1, 3):
                if not any(np.array_equal(pixel, color) for color in target_colors):
                    is_target_color = False
                    break
            
            # 計算外框座標
            x1 = j * tile_width
            y1 = i * tile_height
            x2 = x1 + tile_width
            y2 = y1 + tile_height
            
            # 繪製外框
            if is_target_color:
                cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)  # 藍色外框 (BGR: 255,0,0)
            else:
                cv2.rectangle(image, (x1, y1), (x2, y2), (0, 0, 255), 2)  # 紅色外框 (BGR: 0,0,255)
    
    return image
```

```python
# 定義地板顏色的 BGR 範圍（需根據實際圖片調整）
FLOOR_COLOR_LOW = np.array([90, 120, 100])   # BGR 下限
FLOOR_COLOR_HIGH = np.array([203, 187, 168]) # BGR 上限

def draw_floor_borders(image, all_pixels, rows=6, cols=8, threshold=0.8):
    height, width = image.shape[:2]
    tile_width = width // cols
    tile_height = height // rows
    
    for i in range(rows):
        for j in range(cols):
            grid_pixels = all_pixels[i, j]
            
            # 檢查像素是否在地板顏色範圍內
            mask = cv2.inRange(grid_pixels, FLOOR_COLOR_LOW, FLOOR_COLOR_HIGH)
            match_ratio = np.sum(mask > 0) / (tile_width * tile_height)
            
            # 計算外框座標
            x1 = j * tile_width
            y1 = i * tile_height
            x2 = x1 + tile_width
            y2 = y1 + tile_height
            
            # 根據比例決定外框顏色
            if match_ratio >= threshold:
                cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)  # 藍色（地板）
            else:
                cv2.rectangle(image, (x1, y1), (x2, y2), (0, 0, 255), 2)  # 紅色（非地板）
    
    return image
```

```python

def get_body_region(template_img, body_ratio=0.5):
    h = template_img.shape[0]
    body_start = int(h * body_ratio)
    return template_img[body_start:, :, :]  # 只取身體部分

def get_main_color(img, ignore_white=True):
    reshaped = img.reshape(-1, 3)
    if ignore_white:
        reshaped = np.array([p for p in reshaped if not np.all(p > 230)])
    if reshaped.size == 0:
        return np.array([0, 0, 0])
    return np.mean(reshaped, axis=0)

def detect_character_all_directions(image, character_image_folder, color_tolerance=35):
    image_copy = image.copy()
    image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    for file in os.listdir(character_image_folder):
        char_path = os.path.join(character_image_folder, file)
        template = cv2.imread(char_path)
        if template is None:
            continue

        # 只取下半部（身體）
        body = get_body_region(template)
        main_color_bgr = get_main_color(body)
        main_color_hsv = cv2.cvtColor(np.uint8([[main_color_bgr]]), cv2.COLOR_BGR2HSV)[0][0]

        # HSV容差範圍
        lower_bound = np.clip(main_color_hsv - color_tolerance, 0, 255)
        upper_bound = np.clip(main_color_hsv + color_tolerance, 0, 255)

        # 建立 mask
        mask = cv2.inRange(image_hsv, lower_bound, upper_bound)

        # 形態學操作：清除雜訊
        kernel = np.ones((3, 3), np.uint8)
        mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel, iterations=1)

        # 找輪廓
        contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        for cnt in contours:
            x, y, w, h = cv2.boundingRect(cnt)
            if 10 < w < 50 and 20 < h < 60:  # 根據角色大概尺寸過濾
                body_y = y + h // 2
                cv2.rectangle(image_copy, (x, body_y), (x + w, y + h), (0, 0, 0), 2)

    return image_copy
```

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

# 地板顏色區間（BGR）
FLOOR_COLOR_LOW = np.array([90, 120, 100])
FLOOR_COLOR_HIGH = np.array([203, 187, 168])

# 取得所有 grid 中的 pixels
def get_all_pixels_in_grid(image, rows=6, cols=8):
    height, width = image.shape[:2]
    tile_width = width // cols
    tile_height = height // rows
    grid_pixels = np.zeros((rows, cols, tile_height, tile_width, 3), dtype=np.uint8)

    for i in range(rows):
        for j in range(cols):
            x_start = j * tile_width
            x_end = (j + 1) * tile_width
            y_start = i * tile_height
            y_end = (i + 1) * tile_height
            grid_pixels[i, j] = image[y_start:y_end, x_start:x_end]
    
    return grid_pixels

# 地板與 grid 框線畫出
def draw_floor_borders(image, all_pixels, character_grid=None, exit_grid=None, rows=6, cols=8, threshold=0.8):
    height, width = image.shape[:2]
    tile_width = width // cols
    tile_height = height // rows
    
    for i in range(rows):
        for j in range(cols):
            grid_pixels = all_pixels[i, j]
            mask = cv2.inRange(grid_pixels, FLOOR_COLOR_LOW, FLOOR_COLOR_HIGH)
            match_ratio = np.sum(mask > 0) / (tile_width * tile_height)

            x1, y1 = j * tile_width, i * tile_height
            x2, y2 = x1 + tile_width, y1 + tile_height

            if character_grid == (i, j):
                color = (0, 0, 0)  # 黑色框（角色）
            elif exit_grid == (i, j):
                color = (0, 255, 0)  # 綠色框（出口）
            elif match_ratio >= threshold:
                color = (255, 0, 0)  # 藍色框（地板）
            else:
                color = (0, 0, 255)  # 紅色框（非地板）

            cv2.rectangle(image, (x1, y1), (x2, y2), color, 2)
    
    return image

# 取得 exit 所在格（以模糊 + 中央顏色平均判斷）
# def detect_exit_tile(image, all_pixels, exit_template_path, rows=6, cols=8):
#     template = cv2.imread(exit_template_path)
#     template_blur = cv2.GaussianBlur(template, (5, 5), 0)
#     template_mean = np.mean(template_blur, axis=(0, 1))

#     min_diff = float('inf')
#     target_coord = None

#     for i in range(rows):
#         for j in range(cols):
#             grid = all_pixels[i, j]
#             blur_grid = cv2.GaussianBlur(grid, (5, 5), 0)
#             grid_mean = np.mean(blur_grid, axis=(0, 1))
#             diff = np.linalg.norm(template_mean - grid_mean)

#             if diff < min_diff:
#                 min_diff = diff
#                 target_coord = (i, j)

#     return target_coord

def get_sorted_color_vector(img, center_size=60):
    h, w = img.shape[:2]
    cx, cy = w // 2, h // 2
    half = center_size // 2

    # 擷取中心區塊
    center_patch = img[cy - half:cy + half, cx - half:cx + half]

    # 壓平成 (N, 3)，再排序後攤平成向量
    pixels = center_patch.reshape(-1, 3)
    sorted_pixels = np.sort(pixels, axis=0)
    return sorted_pixels.flatten()


def detect_exit_tile(image, all_pixels, exit_template_path, rows=6, cols=8):
    template = cv2.imread(exit_template_path)
    template_blur = cv2.GaussianBlur(template, (5, 5), 0)
    template_vector = get_sorted_color_vector(template_blur, center_size=60)

    min_diff = float('inf')
    target_coord = None

    for i in range(rows):
        for j in range(cols):
            grid = all_pixels[i, j]
            blur_grid = cv2.GaussianBlur(grid, (5, 5), 0)
            grid_vector = get_sorted_color_vector(blur_grid, center_size=60)

            # 使用 L2 norm 比較距離
            diff = np.linalg.norm(template_vector - grid_vector)

            if diff < min_diff:
                min_diff = diff
                target_coord = (i, j)

    return target_coord


# 角色辨識（多方向角色圖 → 色彩找出最像身體的 grid）
def get_body_region(template_img, body_ratio=0.5):
    h = template_img.shape[0]
    body_start = int(h * body_ratio)
    return template_img[body_start:, :, :]

def get_main_color(img, ignore_white=True):
    reshaped = img.reshape(-1, 3)
    if ignore_white:
        reshaped = np.array([p for p in reshaped if not np.all(p > 230)])
    if reshaped.size == 0:
        return np.array([0, 0, 0])
    return np.mean(reshaped, axis=0)

def detect_character_tile(image, all_pixels, character_image_folder, rows=6, cols=8, color_tolerance=35):
    image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    best_match = None
    max_score = 0

    for file in os.listdir(character_image_folder):
        char_path = os.path.join(character_image_folder, file)
        template = cv2.imread(char_path)
        if template is None:
            continue

        body = get_body_region(template)
        main_color_bgr = get_main_color(body)
        main_color_hsv = cv2.cvtColor(np.uint8([[main_color_bgr]]), cv2.COLOR_BGR2HSV)[0][0]

        lower = np.clip(main_color_hsv - color_tolerance, 0, 255)
        upper = np.clip(main_color_hsv + color_tolerance, 0, 255)

        for i in range(rows):
            for j in range(cols):
                tile = all_pixels[i, j]
                tile_hsv = cv2.cvtColor(tile, cv2.COLOR_BGR2HSV)
                mask = cv2.inRange(tile_hsv, lower, upper)
                score = np.sum(mask > 0)

                if score > max_score:
                    max_score = score
                    best_match = (i, j)

    return best_match

# 主執行區
def run_detection(image, character_image_folder, exit_image_path, rows=6, cols=8):
    # image = cv2.imread(image_path)
    all_pixels = get_all_pixels_in_grid(image, rows, cols)

    exit_grid = detect_exit_tile(image, all_pixels, exit_image_path, rows, cols)
    character_grid = detect_character_tile(image, all_pixels, character_image_folder, rows, cols)

    result = draw_floor_borders(image, all_pixels, character_grid=character_grid, exit_grid=exit_grid, rows=rows, cols=cols)

    # 顯示結果
    plt.figure(figsize=(10, 8))
    plt.imshow(result)
    plt.title("Detected Floor, Character (Black), Exit (Green)")
    plt.axis("off")
    plt.show()
```
